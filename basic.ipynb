```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datax2 = pd.read_csv('data/datax2')
datax2.head()
```

```python
datax3 = pd.read_csv('data/datax3')
datax3.head()
```

```python
x2Train = datax2.values[:,0]
y2Train = datax2.values[:,1]
```

```python
def gradientDescent(x,y):
    w = 0
    m = len(x2Train)
    costList = np.array([])
    iterations = range(1000)
    learningRate = 1e-9
    for i in iterations:
        cost = (0.5/m)*np.sum(np.square(w*x**2-y))
        costList = np.append(costList,cost)
        dw = (1/m)*(np.dot(w*x**2-y,x.T))
        w = w - learningRate*dw
        if i%100==0:print(f'{i}. Cost : {cost}')
    print(f'Value of w : {w}')
    fig,axes = plt.subplots(1,2,figsize=(12,6))
    axes[0].plot(costList)
    axes[0].set_title('Cost Function')
    axes[0].set_ylabel('y')
    axes[1].scatter(x,y,label='Orignal')
    axes[1].plot(x,w*x**2,label='Predicted',c='r')
    axes[1].legend()
    fig.tight_layout()
```

```python
gradientDescent(x2Train,y2Train)
```
