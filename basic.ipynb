```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datax2 = pd.read_csv('data/datax2')
datax2.head()
```

```python
datax3 = pd.read_csv('data/datax3')
datax3.head()
```

```python
x2Train = datax2.values[:,0]
y2Train = datax2.values[:,1]
x3Train = datax3.values[:,0]
y3Train = datax3.values[:,1]
```

```python
def gradientDescent(x,y,iterations,learningRate,power):
    m = x.size
    w = 0
    costList = np.array([])
    predict = np.dot(x**power, w)
    for i in range(iterations):
        cost = (1/(2*m))*np.sum(np.square(predict)-y)
        costList = np.append(costList,cost)
        dw = (1/m)*np.dot(x.T,predict-y)
        w = w-learningRate*dw
        if(i%(iterations/10)==0):print(f'{i}. Cost ={cost}')
    return w,costList
```

```python
w,costList = gradientDescent(x2Train,y2Train,1000,1e-9,2)
fig,axes = plt.subplots(1,2,figsize=(8,4))
axes[0].scatter(x2Train,y2Train,label='Training Data',marker=".",c='r')
axes[0].plot(x2Train, np.dot(x2Train**2, w), label='Predicted Data')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Value')
axes[0].legend()
axes[1].plot(costList)
axes[1].set_title('Cost Function')
fig.tight_layout()
```
